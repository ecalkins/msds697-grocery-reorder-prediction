```python
import pandas as pd
```

```python
orders = pd.read_csv('orders.csv')
prods = pd.read_csv('products.csv')
depts = pd.read_csv('departments.csv')
aisles = pd.read_csv('aisles.csv')

orders_train = pd.read_csv('order_products__train.csv')
orders_prior = pd.read_csv('order_products__prior.csv')

df = pd.concat((orders_train, orders_prior), axis=0)

df = orders.merge(df, on='order_id', how='left')
df = df.merge(prods, on='product_id', how='left')
df = df.merge(depts, on='department_id', how='left')
df = df.merge(aisles, on='aisle_id', how='left')
```
